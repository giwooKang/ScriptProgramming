다형성에 대해 설명하고 다형성을 보여주는 자신만의 파이썬 코드 예제를 제시하시오.

<font color=blue>
다형성은 흔히 알려진 메소드의 타입별 구현이라는 단순한 의미에 그치지 않는 포괄적인 개념이다. 다형성은 객체만이 가질 수 있는 것이 아니며 상수, 변수, 식, 오브젝트, 함수, 메소드 등, 코드의 각 단위 요소들도 갖는다. 흔히 단형성 언어에서 이들이 '불변하는 단일한 의미'를 갖는 것에 반해, 다형성 언어에서는 '가변할 수 있는 다양한 의미'를 갖는다. <br/><br/>
하지만 흔히 포괄적인 개념에 비해서는 단순한 활용을 하게 되는데 이것이 흔히 다형성인것으로 착각하는 방법, 즉 동일한 객체의 동일한 메소드가 여러 종류의 파라미터를 받고 그에 대응하는 여러 종류의 결과값을 낼 수 있는 기능을 구현하는 것이다. 예로 아래 코드를 보자.


</font>

```python
def floatToString(f):
    return str(f)
def intToString(i):
    return str(i)

print floatToString(3.1)
print intToString(3)
```

<font color=blue>
파이썬에서는 변수의 Type이 동적으로 결정되므로 단형성 언어를 표현하기에는 조금 불편하다. 따라서 메서드의 파라미터인 f는 반드시 실수 자료형만 받을 수 있고, i는 반드시 정수 자료형만 받을 수 있다고 가정하자. 그렇다면 위의 floatToString 메서드와 intToString 메서드는 오직 한 종류의 파라미터를 갖고 하나의 결과값을 갖는 단형성 메서드가 된다. 각 메서드는 이 기능 외에 다른 방법으로는 사용할 수 없다. 반면 아래 코드를 보자.
</font>

```python
class cData:
    def __init__(self, data):
        self.data = data
    def toString(self):
        return str(self.data)
        
class cFloat(cData):
    def toString(self):
        if type(self.data) is float:
            return str(self.data)
        else:
            return "실수형 데이터가 아닙니다."
        
class cInt(cData):
    def toString(self):
        if type(self.data) is int:
            return str(self.data)
        else:
            return "정수형 데이터가 아닙니다."

for i in (cData(3), cFloat(3.1), cFloat(3), cInt(3), cInt(3.1)):
    if isinstance(i, cData):
        print i.toString()
    else:
        print "이 객체는 cData 타입이 아닙니다."
```

<font color=blue>
파이썬 특유의 유연한 구조는 엄격한 타입 검사를 수행하는 코드를 구현하려면 되려 기존보다 코드가 늘어나는 단점 아닌 단점이 있다... 위의 코드를 보면 다형성을 보이기 위해 선언한 3개의 클래스를 볼 수 있다. cData는 특정 데이터를 wrapping하는 클래스이고, cFloat과 cInt는 cData를 상속한다. 세 클래스 모두 toString() 메서드를 구현한다. 같은 이름이지만 동작은 모두 다르다.<br/><br/>
출력 부분의 코드를 보면 for 반복문에 넘어간 튜플 내의 모든 원소가 cData 클래스의 인스턴스라는 것을 확인할 수 있다. 이 for 반복문은 튜플의 각 원소에 대해 toString을 수행한다. 단형성 언어라면 toString() 메서드의 결과는 모두 똑같아야 한다. <br/><br/>이제 출력된 결과를 확인해보자. cFloat의 toString은 생성자에서 얻은 data가 실수형이 아니라면 에러 메시지를 출력한다. 반면 cInt는 data가 정수형이 아니면 에러 메시지를 출력한다. 튜플의 원소가 모두 cData형인데, 실행되는 toString()의 작동이 모두 제각각이다. 이렇게 각 객체가 같은 메서드를 서로다른 방법으로 동작하도록 하는 방법으로 흔히 다형성이 응용된다. 정리하면, 코드의 구성 요소중 하나인 메서드(이 경우에는 toString())가 하나의 단일한 의미를 갖는 것이 아닌, 다양한 의미를 갖도록 하는 것이 바로 다형성이다.

 > \>\>\> c()  <br/>
\>\>\> d()  <br/>
\>\>\> print c  <br/>
[Count (step: 1)] 12  <br/>
\>\>\> print d  <br/>
[Count (step: 2)] 14  <br/>

* 요구사항 5. 다음과 같은 두 개의 산술 연산 (+, -)이 수행될 수 있도록 Counter 클래스 내에 관련 메소드를 추가하시오.

 > \>\>\> c = c + 5  <br/>
\>\>\> d = d - 5  <br/>
\>\>\> print c  <br/>
[Count (step: 1)] 17  <br/>
\>\>\> print d  <br/>
[Count (step: 2)] 9  <br/>

* 요구사항 6. 다음과 같은 관계연산 (>, <, ==)이 수행될 수 있도록 Counter 클래스 내에 관련 메소드를 추가하시오.

 > \>\>\> print c > 10  <br/>
True  <br/>
\>\>\> print d > 10  <br/>
False  <br/>
\>\>\> print c < 10  <br/>
False  <br/>
\>\>\> print d < 10 <br/> 
True  <br/>
\>\>\> print c == 17  <br/>
True  <br/>
\>\>\> print d != 9  <br/>
False  <br/>

```python
class Counter:
    # 요구사항 1 구현
    def __init__(self, initValue, delta = 1):
        self.value = initValue
        self.delta = delta
        
    # 요구사항 2 구현
    def __str__(self):
        return "[Count (step: " + str(self.delta) + ")] " + str(self.value)
    
    # 요구사항 3 구현
    def incr(self):
        self.value += delta
    
    # 요구사항 4 구현
    def __call__(self):
        self.incr()
    
    # 요구사항 5 구현
    def __add__(self, temp):
        return self.value + temp
    
    def __sub__(self, temp):
        return self.value - temp
    
    # 요구사항 6 구현
    def __cmp__(self, temp):
        return self.value - temp
```

Counter 의 구현은 클래스 사용법에 관련된 내용으로 크게 어렵지 않게 구현가능하다. __ 으로 시작하고 끝나는 메소드명은 예약어로, 해당 클래스 타입에 연산자 오버로딩을 추가하거나, 특정 처리에 대응하여 처리할 코드를 추가하기 위해 사용한다.

<h3>문제 4)</h3> 
<br/>
다음은 내장 자료형 list를 서브클래싱하여 만든 MySet 클래스 정의 내용이다. 다음 클래스 정의에서 __init__(), __str()__(), elimicate_duplicate()의 세 개의 메소드 코드 내용을 자신이 다른 사람에게 가르친다고 생각하며 설명해보시오.

* MySet은 집합(Set) 자료형을 정의하려는 의도하에 만들어진 클래스이다.

```python
class MySet(list):
    def __init__(self, l):
        for e in l:
            self.append(e)
        MySet.eliminate_duplicate(self)
    
    def __str__(self):
        result = "MySet: {"
        for e in self:
            result = result + str(e) + ", "
        result = result[0:len(result)-2] + "}"
        return result

    @staticmethod    
    def eliminate_duplicate(l):
        s = []
        for e in l:
            if e not in s:
                s.append(e)
        l[:] = []
        for e in s:
            l.append(e)

    
if __name__ == "__main__":
    s = MySet([1, 2, 2, 3])
    print s
    t = MySet([2, 3, 4, 5, 6, 7, 8, 8, 8, 8, 8, 9])
    print t
```

<font color=blue>
\_\_init\_\_ 는 생성자 코드로, 임의의 리스트 l을 파라미터로 받는다. MySet의 인스턴스가 생성될 때, l의 모든 원소를 자신에게 추가(MySet은 list을 상속) 한 뒤 eliminate\_duplicate를 호출한다. <br/><br/> 
\_\_str\_\_ 는 print, 혹은 str() 등 객체를 문자열화 할 필요가 있을 때 수행되는 메서드이다. 여기서는 "MySet: { [listitem, lisitem, ...] }" 형태의 출력 포맷에 맞게 MySet 인스턴스를 문자열화하는 코드가 추가되어 있다.
<br/><br/>
eliminate\_duplicate는 중복 제거라는 이름에 맞게, 리스트 내의 각 요소에 대해 중복되는 요소를 제거하는 역할을 한다. 새로운 리스트를 만들고, 이 리스트에는 MySet의 인스턴스가 갖고 있는 리스트 원소를 추가한다. 그러나 만약 복사하려는 원소를 이미 포함하고 있다면 추가하지 않는다. 이후에 MySet의 인스턴스가 갖고 있는 리스트를 초기화하고, 중복 제거된 리스트의 원소들을 다시 추가한다.

<h3>문제 5)</h3> 
<br/>
4번 문제에 정의된 MySet 클래스에 메소드를 추가하여 다음 각 요구사항 모두를 만족시키는 코딩을 제시하시오

* 정답을 각 요구사항별로 입력할 필요 없이 요구사항 3개 전체에 대해 1개의 MySet 클래스 정의 코드를 제시하고 주석등으로 해당 코드를 설명한다.
* 요구사항 1. | 연산으로 두 집합의 합집합을 반환한다.

 > \>\>\> u = s | t <br/>
\>\>\> print u  <br/>
MySet: {1, 2, 3, 4, 5, 6, 7, 8, 9}  <br/>
 
* 요구사항 2. & 연산으로 두 집합의 교집합을 반환한다.

 > \>\>\> u = s & t  <br/>
\>\>\> print u  <br/>
MySet: {2, 3}  <br/>

* 요구사항 3. - 연산으로 두 집합의 차집합을 반환한다.

 > \>\>\> s = MySet([1, 2, 3])  <br/>
\>\>\> t = MySet([3, 4, 5])  <br/>
\>\>\> u = s - t  <br/>
\>\>\> print u  <br/>
MySet: {1}  <br/>

```python
class MySet(list):
    # 원본 코드
    def __init__(self, l):
        for e in l:
            self.append(e)
        MySet.eliminate_duplicate(self)
    
    def __str__(self):
        result = "MySet: {"
        for e in self:
            result = result + str(e) + ", "
        result = result[0:len(result)-2] + "}"
        return result

    @staticmethod    
    def eliminate_duplicate(l):
        s = []
        for e in l:
            if e not in s:
                s.append(e)
        l[:] = []
        for e in s:
            l.append(e)
            
    # 요구사항 1 구현
    def __or__(self, l):
        s = []
        for e in l + self:
            s.append(e)
        MySet.eliminate_duplicate(s)
        return MySet(sorted(s))
        
    # 요구사항 2 구현
    def __and__(self, l):
        s = []
        for e in l:
            if e in self:
                s.append(e)
        return MySet(s)
    
    # 요구사항 3 구현
    def __sub__(self, l):
        s = []
        for e in self:
            if not e in l:
                s.append(e)         
        return MySet(s)
```

<font color=blue>
\_\_or\_\_ 메서드는 self, l 파라미터를 받아 이 두 리스트의 모든 요소를 s에 넣고, s에 대해 중복제거를 수행한 뒤 정렬한다. 이 리스트를 이용해 MySet 인스턴스를 만들어 리턴한다. 즉, 합집합을 구현한다.<br/><br/>
\_\_and\_\_ 메서드는 self, l 파라미터를 받아 l 리스트를 돌면서 self에도 해당 요소가 있으면 s에 추가한다. 즉, 교집합을 구현한다.<br/><br/>
\_\_sub\_\_ 메서드는 self 의 요소를 순회하면서 l에 해당 요소가 없는 경우에만 s에 추가한다. 즉, 차집합을 구현한다.<br/><br/>

<h3>문제 7 - Project Euler 16번)</h3> 
<br/>
2^15 = 32768 의 각 자리수를 더하면 3 + 2 + 7 + 6 + 8 = 26 입니다.

2^1000의 각 자리수를 모두 더하면 얼마입니까?

```python
s = 1
for i in range(1, 1001):
    s = s * 2

s_str = str(s)
sum = 0
for ch in range(0, len(s_str)):
    sum += int(s_str[ch])
    
print sum
```

<font color=blue>
1부터 1000까지 돌면서 s에 2를 곱해 2^1000을 얻고, 이를 string으로 만든다. string의 길이만큼 for를 돌면서 각 자리의 숫자를 int로 변환한 값을 sum에 누적한다. 답을 매우 쉽게 얻을 수 있다.

1부터 1,000까지 영어로 썼을 때는 모두 몇 개의 글자를 사용해야 할까요?

참고: 빈 칸이나 하이픈('-')은 셈에서 제외하며, 단어 사이의 and 는 셈에 넣습니다.<br/>
* 예를 들어 342를 영어로 쓰면 three hundred and forty-two 가 되어서 23 글자,<br/>
115 = one hundred and fifteen 의 경우에는 20 글자가 됩니다.

```python
# zero one two three four five six seven eight nine
# 0의 경우는 읽지 않는다.
place_units = [0, 3, 3, 5, 4, 4, 3, 5, 5, 4]
# 10의 자리의 경우, 값이 1인 경우와 그렇지 않은 경우에 다르다.
place_ten_if_1 = [3, 6, 6, 8, 8, 7, 7, 9, 8, 8]
place_ten_else = [0, 0, 6, 6, 5, 5, 5, 7, 6, 6]
# 100의 자리의 경우 units 에 hundred 가 붙고, 1000의 자리의 경우 units 에 thousand 가 붙는다. 즉 +7, +8 로 해결할 수 있다.

def numStrCounter(num):
    s_num = str(num)
    numStrCount = 0
    i = 0
    
    # 1000의 자리 계산
    if len(s_num)==4:
        return 11
        
    # 100의 자리 계산
    if len(s_num)>=3:
        numStrCount += place_units[int(s_num[i])] + 7
        
        # and 추가 여부 계산
        if not s_num[i+1 : len(s_num)] == "00":
            numStrCount+=3    
        i += 1    
        
    # 10의 자리 계산
    if len(s_num)>=2:
        if int(s_num[i]) == 1:
            numStrCount += place_ten_if_1[int(s_num[i+1])]
            return numStrCount
        else:
            numStrCount += place_ten_else[int(s_num[i])]
            i += 1
            
    # 1의 자리 계산
    if len(s_num)>=1:
        numStrCount += place_units[int(s_num[i])]
        
    return numStrCount

sum = 0
for i in range(1, 1001):
    sum += numStrCounter(i)
    
print sum
```

```python
import urllib2
import operator
import string

# 1) WebWordsFrequency 정의
class WebWordsFrequency:
    # 기존 코드
    def removeWebCode(self, source, start, end):
        str_var = ""
        subList = source.split(start)

        for i in range(0, len(subList)):
            sub = subList[i]
            index = sub.rfind(end)
            if index == -1 or index == len(sub)-1:
                if i!=0 and i!=len(subList)-1:
                    continue
            str_var += sub[index+1:] + " "
        str_var = " ".join(str_var.split())
        return str_var

    def getWordsFrequencyBySingleUrl(self, url):
        source = urllib2.urlopen(url).read()       
        source = self.removeWebCode(source, '<script', '/script>')
        source = self.removeWebCode(source, '<style', '/style>')
        source = self.removeWebCode(source, '<', '>')

        for i in range(0, len(string.punctuation)):
            source = source.replace(string.punctuation[i], " ")

        words = {}
        for i in source.split():
            if i in stopWords:
                continue
            if not i in words.keys():
                words[i]=1
            else:
                words[i]=words[i]+1
                
        return words

    # Assignment_5 코드
    
    # 2) url을 임의의 개수 넣을 수 있는 생성자 정의
    def __init__(self, *arg):
        self.urlList = []
        for url in arg:
            if not url in self.urlList:
                self.urlList.append(url)
    
    # 3) addUrl을 이용해 url을 단 하나 추가할 수 있음
    def addUrl(self, url):
        if not url in self.urlList:
            self.urlList.append(url)
    
    # 4) removeUrl을 이용해 url을 단 하나 삭제할 수 있음
    def removeUrl(self, url):
        if url in self.urlList:
            self.urlList.remove(url)
    
    # 5) listUrls() 메소드 구현
    def listUrls(self):
        for url in self.urlList:
            print url
    
    # 6) getWordsFrequency 메소드 구현
    def getWordsFrequency(self):
        words = {}
        for url in self.urlList:
            wordsInUrl = self.getWordsFrequencyBySingleUrl(url)
            for key in wordsInUrl.keys():
                if not key in words.keys():
                    words[key] = wordsInUrl[key]
                else:
                    words[key] += wordsInUrl[key]
        return words

    # 7) getMaxFreqencyWords 메소드 구현
    def getMaxFreqencyWords(self):
        words = self.getWordsFrequency()
        
        if not len(words.items()) == 0:
            wordList = []
            sortedTupleList = sorted(words.items(), key=operator.itemgetter(1), reverse=True)
            freq = sortedTupleList[0][1] 
            i = 0
            while sortedTupleList[i][1] == freq:
                wordList.append(sortedTupleList[i][0])
                i+=1
                if i == len(sortedTupleList):
                    break
            return wordList
        else:
            return None
        
    # 8) searchUrlByWord 메소드 구현
    def searchUrlByWord(self, word):
        wordsList = []
        max = -1
        for url in self.urlList:
            wordsInUrl = self.getWordsFrequencyBySingleUrl(url)
            if word in wordsInUrl.keys():
                wordsList.append((url, wordsInUrl[word]))
                if wordsInUrl[word] > max:
                    max = wordsInUrl[word]
        if not max == -1:
            for tup in wordsList:
                if max == tup[1]:
                    print tup[0]
        else:
            print word, "is not exist any site"
```

<h3>문제 12)</h3> 
<br/>
다음과 같은 코딩이 가능하도록 OrderedWebWordsFrequency 안에 반복자와 관련된 메소드를 추가하시오.

> \>\>\> for i in w4:<br/>
\>\>\>       print i<br/>
('site, 12)<br/>
('science', 11)<br/>
('hello', 8)<br/>
('world', 2)<br/>
('program', 1)<br/>
('python', 1)

```python
class OrderedWebWordsFrequency(WebWordsFrequency):
    # 11번 문제
    def getWordsFrequency(self, reverse=False):
        w1 = WebWordsFrequency()
        for url in self.urlList:
            w1.addUrl(url)
        words = w1.getWordsFrequency()
        sortedWordsItem = sorted(words.items(), key=operator.itemgetter(1), reverse=not reverse)
        return sortedWordsItem
    
    # 12번 문제
    def __iter__(self):
        return iter(self.getWordsFrequency())
```

```python
w4 = OrderedWebWordsFrequency('http://www.times.com/', 'https://www.amazon.com/', 'https://github.com/')
print w4.getWordsFrequency()
```

```python
print w4.getWordsFrequency(reverse=True)
```

12번 문제는 별다른 조건이 없었으므로, 11번의 결과를 iter() 로 캐스팅하여 반환한다. 단 한 줄이면 된다.
