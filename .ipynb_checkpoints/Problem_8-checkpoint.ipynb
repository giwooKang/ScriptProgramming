문제에서 주어진 데이터를 가공하지 않고 그대로 사용하려고 했으나 왜인지 더 손 댈 부분이 많아진 것 같은 코드이다. 어떤 방법을 쓰던, 1000개의 문자를 갖는 문자열 str를 정의한다.

```python
max = -1
for i in range(0, 996):
    acc = 1
    for j in str[i:i+5]:
        acc = acc * int(j)
    if acc > max:
        max = acc
print max
```

역시 매우 간단한 코드로, i가 0부터 996까지 돌면서 i~(i+5)의 배열 위치를 slice한다. 또한 slice된 배열을 다시 foreach로 돌면서 각 요소를 int로 캐스팅하여 곱한 값을 누적한다. 이 값이 최대값이라면 이를 최대값 max 변수에 반영한다. 반복이 끝나면 max값을 리턴한다.
