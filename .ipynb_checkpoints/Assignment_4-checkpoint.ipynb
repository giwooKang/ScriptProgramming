- 3) 파일 s.txt을 읽어서 각 라인들에 있는 모든 단어들을 순차적으로 다시 나열하되 각 라인에 세 개의 단어들이 오도록 하여 s3.txt에 기록하는 코드를 작성하시오. 즉, 프로그램 수행 후 s3.txt에 있는 파일 내용은 다음과 같다.

pig ham cat<br/>

```python
# 파일 생성
f = open('s.txt', 'w')
f.write("""pig ham
cat dog
ham bird
dog pig""")
f.close()
```

```python
# 1) 
def fileReader(name):
    f = open(name, 'r')
    L = []
    for line in f:
        line = line.replace('\n', '')
        L.append(line)
    return L

def fileWritter(name, L):
    fout = open(name, 'w')
    fout.write('\n'.join(L))
    fout.close()
    
L = fileReader('s.txt')
L.sort() 
fileWritter('s1.txt', L)
```

```python
# 2) 
def secondWord(str):
    str = str.split(' ')
    return str[1]

L = fileReader('s.txt')
L.sort(key=secondWord)
fileWritter('s2.txt', L)
```

```python
# 3) 
L = fileReader('s.txt')
str = ""
for line in L:
    str += line + " "
    
fout = open('s3.txt', 'w')
i = 0
for line in str.split(' '):
    fout.write(line + ' ')
    i += 1
    if i%3 == 0:
        fout.write('\n')
    
fout.close()
```

```python
def printFile(name):
    f = open(name, 'r')
    for line in f:
        print line.replace('\n', '')

print "string in file [s.txt] is..."
printFile('s.txt')

print "\nstring in file [s1.txt] is..."
printFile('s1.txt')

print "\nstring in file [s1.txt] is..."
printFile('s2.txt')

print "\nstring in file [s1.txt] is..."
printFile('s3.txt')
```

<h3>문제 3)</h3> 
<br/>
range() 함수와 유사한 frange() 함수를 다음 조건을 참고하여 만드시오.

* 1) frange()의 인자 구성은 다음과 같이 range와 동일하지만 각 인수들은 음수를 받지 않는다고 가정한다.

        range(stop)
        range(start, stop[, step])
        
* 2) frange() 함수의 인자에 대한 기본 시작(start) 값은 0.0이고, 기본 단계(step) 값은 0.1이다.
* 3) frange 사용 예
    * 3-1) frange(0.5)
        * [0.0, 0.1, 0.2, 0.3, 0.4]
    * 3-2) frange(1.0, 2.0)
        * [1.0, 1.1, 1.2, 1.3, 1.4, 1.5, 1.6, 1.7, 1.8, 1.9]
    * 3-3) frange(2.2, 4.0, 0.5)
        * [2.2, 2.7, 3.2, 3.7]

```python
def frange_method(start, end, step=0.1):
    L=[]
    i = start
    while(i < end):
        L.append(round(i, 10))
        i += step
    return L       

def frange(*arg):
    if len(arg)==1:
        return frange_method(0.0, arg[0])
    elif len(arg)==2:
        return frange_method(arg[0], arg[1])
    elif len(arg)==3:
        return frange_method(arg[0], arg[1], arg[2])
    
print frange(0.5)
print frange(1.0, 2.0)
print frange(2.2, 4.0, 0.5)
```

<font color=blue>파이썬에서는 함수 오버로딩이 따로 지원되지 않아서 여러 파라미터를 동시에 받을 수 있는 방법을 이용했다. 입력되지 않은 파라미터의 초기값을 지정하고, 파라미터의 개수에 따라 다르게 함수를 호출하면 간단히 구현 가능하다.
<br/><br/> 
파이썬의 Decimal 모듈을 이용하면 부동소수점 연산을 완벽히 할 수 있지만, 입력 데이터 자체를 string으로 주지 않는 이상 원본 float 실수 값 자체가 부동소수점 에러를 갖고 있기 때문에 실수 값을 파라미터로 받아 처리하기 위해서는 적당한 오차 범위 뒤에서 반올림을 해 주는 것이 구현이 쉽다.</font>

<h3>문제 4)</h3> 
<br/>
가변인수를 받는 함수 sum()을 다음과 같은 조건을 참고하여 구현하시오

* sum() 사용 예
    * sum()
        * 0
    * sum(1, 2)
        * 3
    * sum(1, 2, 3, 4, 5)
        * 15
    * sum(1, 5, 7, 2, -10)
        * 5

```python
def sum(*arg):
    s = 0;
    for num in arg:
        s += num
    return s

print sum()
print sum(1, 2)
print sum(1, 2, 3, 4, 5)
print sum(1, 5, 7, 2, -10)
```

<font color=blue>*arg 를 통해 가변인수를 파라미터로 받을 수 있다. 이후 연산은 크게 어렵지 않다.</font>

<h3>문제 5)</h3> 
<br/>
여러 단어로 이루어진 문자열을 입력받아 각 단어의 첫글자로 이루어진 단어를 대문자로 출력하는 myinitial() 함수를 다음 조건을 참고하여 작성하시오.

* 1) 다음에 제시되는 함수들을 모두 이용해야 한다.
    * split
    * map
    * join
* 2) myinitial() 함수 사용 예
    * myinitial("as soon as possible")
        * ASAP

```python
def returnFirstCharUpper(s):
    return s[0].upper()

def myinitial(str):
    return "".join(map(returnFirstCharUpper, str.split(' ')))

print myinitial("as soon as possible")
```

<font color=blue>returnFirstCharUpper는 특정 문자열의 맨 앞 글자를 대문자로 바꿔서 반환하는 함수이다. 이를 map에 입력 str를 split(' ')한 결과와 함께 파라미터로 넘기면 각 단어의 앞 문자만 남아 대문자로 바뀐 리스트를 얻을 수 있다. 이를 "".join 으로 합쳐서 출력하면 myinitial을 구현할 수 있다. </font>

<h3>문제 6)</h3> 
<br/>
음이 아닌 정수 n를 입력받으면 n! (factorial)을 계산하는 myfact() 함수를 재귀적 함수로 구현하시오.

* [옵션] 참을 수 있는 정도 만큼의 수행시간을 직접 기다려보면서 n을 늘려보도록 합시다.

```python
import sys
sys.setrecursionlimit(5000)

def myfact(n):
    if(n==1):
        return 1
    return n * myfact(n-1)

print myfact(4467)
```

<font color=blue>[옵션] 항목이 좀 의아하긴 한데, 어쨋든 구현해 보았다. 현재 테스팅 환경에서는 4467 까지 값을 출력하고, 이후부터는 파이썬 커널이 작동을 중지하면서 튕긴다. 
<br/><br/>
문제의 의도는 maximum recursion depth exceeded 에러를 확인하라는 것 같긴 한데, 또 수행시간을 기다리라는 것을 봐서는 그런 것도 아닌 듯 하다. 아무래도 애매한 옵션이다. 후자이기도 애매한 것이 고작 5천번 정도는 순식간에 계산이 끝나기 때문인데... 어떤걸 의도한 것인지 잘 모르겠다.

<h3>문제 8 - Project Euler 10번)</h3> 
<br/>
10 이하의 소수를 모두 더하면 2 + 3 + 5 + 7 = 17 이 됩니다.

이백만(2,000,000) 이하 소수의 합은 얼마입니까?

```python
def factorization(num):
    count = pow(num, 0.5) + 1
    i = 2
    if num==2:
        return True
    while i < count:
        if num%i==0:
            return False
        i += 1
    return True

sum = 0
i = 2
while(i<=2000000):
    if factorization(i):
       sum += i
    i+=1
print sum
```

<font color=blue> factorization 함수는 해당 수가 소수인지 아닌지 확인하고 bool 값을 반환한다. 2부터 200만까지 반복하면서 해당 수가 소수이면 더하고, 소수가 아니면 더하지 않는다. 반복문이 끝나면 해당 결과를 반환한다.
</font>

<h3>문제 10 - Project Euler 12번)</h3> 
<br/>
1부터 n까지의 자연수를 차례로 더하여 구해진 값을 삼각수라고 합니다.<br/>
예를 들어 7번째 삼각수는 1 + 2 + 3 + 4 + 5 + 6 + 7 = 28이 됩니다.<br/>
이런 식으로 삼각수를 구해 나가면 다음과 같습니다.

    1, 3, 6, 10, 15, 21, 28, 36, 45, 55, ...
    
이 삼각수들의 약수를 구해봅시다.

     1: 1
     3: 1, 3
     6: 1, 2, 3, 6
    10: 1, 2, 5, 10
    15: 1, 3, 5, 15
    21: 1, 3, 7, 21
    28: 1, 2, 4, 7, 14, 28

위에서 보듯이, 5개 이상의 약수를 갖는 첫번째 삼각수는 28입니다.

그러면 500개 이상의 약수를 갖는 가장 작은 삼각수는 얼마입니까?

```python
def factorization(num, dic):
    count = pow(num, 0.5) + 1
    i = 2
    while i < count:
        if num%i==0:
            if not i in dic.keys():
                dic[i] = 1
            else:
                dic[i] = dic[i] + 1
            num = num/i
            return factorization(num, dic)
        i += 1
    if not num in dic.keys():
        dic[num] = 1
    else:
        dic[num] = dic[num] + 1

def divisorCounter(n):
    dic = {}
    factorization(n, dic)
    sum = 1
    for i in dic.values():
        sum = sum * (i+1)
    return sum

i = 0
t_num = 0
while(1):
    i += 1
    t_num += i
    n_num = divisorCounter(t_num)
    if n_num >= 500:
        print t_num 
        break
```

<font color=blue>
divisorCounter 는 n을 파라미터로 받아 해당 수를 소인수분해하고, 그 약수의 지수를 이용해 해당 수의 약수의 개수를 계산하는 함수이다. 예로, 10은 2^1*5^1 로 소인수분해된다. 각 약수의 지수에 1을 더하여 곱하면 4가 되는데, 이는 10의 약수의 개수와 일치한다. 해당 공식을 이용하면 빠르게 약수의 개수를 계산할 수 있다.<br/><br/>
i를 증가시키고, 삼각수 t_num에 i를 누적시켜나가면서 t_num의 약수를 센다. 약수를 세었을 때, 그 값이 500개를 넘으면 해당 값을 출력하고 종료한다.
</font>

<h3>문제 12 - Project Euler 14번)</h3> 
<br/>
양의 정수 n에 대하여, 다음과 같은 계산 과정을 반복하기로 합니다.

    n → n / 2 (n이 짝수일 때)
    n → 3 n + 1 (n이 홀수일 때)

13에 대하여 위의 규칙을 적용해보면 아래처럼 10번의 과정을 통해 1이 됩니다.

    13 → 40 → 20 → 10 → 5 → 16 → 8 → 4 → 2 → 1
    
아직 증명은 되지 않았지만, 이런 과정을 거치면 어떤 수로 시작해도 마지막에는 1로 끝나리라 생각됩니다.<br/>
(역주: 이것은 콜라츠 추측 Collatz Conjecture이라고 하며, 이런 수들을 우박수 hailstone sequence라 부르기도 합니다)

그러면, 백만(1,000,000) 이하의 수로 시작했을 때 1까지 도달하는데 가장 긴 과정을 거치는 숫자는 얼마입니까?

참고: 계산 과정 도중에는 숫자가 백만을 넘어가도 괜찮습니다.

```python
def collatz(n):
    if n==1:
        return 1
    if n%2==0:
        return 1 + collatz(n/2)
    else:
        return 1 + collatz(3*n+1)

max_count = -1
max = 0
for i in range(1, 1000001):
    count = collatz(i)
    if count > max_count:
        max = i
        max_count = count
        
print max
```

<font color=blue>
collatz 함수는 재귀적으로 해당 연산을 수행하며, 결과적으로 계산 과정을 거친 count를 반환한다. 이를 이용하여 1부터 1000000까지 반복하면서 각 수에 대해 count를 계산하고, 가장 큰 계산 횟수를 갖는 값을 max에 누적한다. 반복문이 끝나면 max를 반환한다.
</font>

```python
def myfact(n):
    if(n==1):
        return 1
    return n * myfact(n-1)

print myfact(40) / (myfact(20) * myfact(20))
```

<font color=blue>
n*n 사각형을 생각할 때, 어느 경로던지간에 해당 경로의 길이가 2n임은 분명하다. 2n상의 경로에서, 각각의 시점에 오른쪽인지, 왼쪽인지를 결정하는 부분이 다를 뿐이다. 2*2의 경우에서 간단히 예를 들면, 2*2 상의 격자에서 경로의 길이는 4이다. 그리고 총 6개의 가지수가 있는데, 각각 (r, r, d, d), (r, d, r, d), (r, d, d, r), (d, r, r, d), (d, r, d, r), (d, d, r, r) 이다. 논리적인 머리가 비상하다면, 이는 한 가지 변수로 나타낼 수 있다. 즉 간다와 가지 않는다이다. (r, r, d, d)는 (r, r, !r, !r) 이다. 이건 더 알기 쉽게 표현해서, 4개의 r 중에서 2개를 뽑는다고 생각할 수 있다. 그렇다. 여기까지 왔다면 Combination이 떠올라야 한다. 4개의 r중 2개를 뽑는다. 순서가 중요치 않으나 중복은 허용하지 않는다. 4 C 2 = 6. 그러므로 위 예제는 6이 된다.<br/><br/>
20 * 20 격자에 대해서 경로의 길이는 20+20, 40이다. 40개의 경로에 대해서 20개를 택하는 조합의 경우의 수는? 40 C 20이다.

```python
import pickle
import urllib2
import string

def removeWebCode(source, start, end):
    str = ""
    subList = source.split(start)

    for i in range(0, len(subList)):
        sub = subList[i]
        index = sub.rfind(end)
        if index == -1 or index == len(sub)-1:
            if i!=0 and i!=len(subList)-1:
                continue
        str += sub[index+1:] + " "
    str = " ".join(str.split())
    return str

def project(url):
    source = urllib2.urlopen(url).read()
    
    # 2) 파일로 웹페이지를 저장하는 코드
    f = open(url.replace('http://', '').replace('https://', '').replace('/', '') + '.html', 'w')
    f.write(source)
    f.close()
    
    # Script code 제거
    source = removeWebCode(source, '<script', '/script>')
    # CSS code 제거
    source = removeWebCode(source, '<style', '/style>')
    # HTML Tag 제거
    source = removeWebCode(source, '<', '>')
        
    for i in range(0, len(string.punctuation)):
        source = source.replace(string.punctuation[i], "")
    
    words = {}
    for i in source.split():
        # 1) 단어 사전에서 영어 불용어 제거하는 코드
        if i in stopWords:
            continue
        if not i in words.keys():
            words[i]=1
        else:
            words[i]=words[i]+1
    
    # 3) 파일에 단어 사전을 저장하는 코드
    f = open(url.replace('http://', '').replace('https://', '').replace('/', '') + '.words.frequency', 'w')
    pickle.dump(words, f)
    f.close()       

# 4) 최소 5개 이상의 영어 웹 사이트 출현빈도 사전 저장
    
# BBC news
project('http://www.bbc.com/')
# time 
project('http://time.com/')
# us news
project('http://www.foxnews.com/')
# abc news
project('http://abcnews.go.com/')
# cnn news
project('http://edition.cnn.com/')
```

```python
# 5) 저장한 모든 words.frequency 파일을 객채로 다시 로드, 가장 많이 출현한 단어 3개 제시
import operator

filename=['www.bbc.com.words.frequency', 'time.com.words.frequency', 'www.foxnews.com.words.frequency', 'abcnews.go.com.words.frequency', 'edition.cnn.com.words.frequency']
file_word_dic = {}

for file in filename:
    f = open(file, 'r')
    dic = pickle.load(f)
    f.close()
    file_word_dic[file] = dic
    sortedList = sorted(dic.items(), key=operator.itemgetter(1), reverse=True)
    print file + " : ",
    for i in range(0, 3):
        print sortedList[i],
    print
```
